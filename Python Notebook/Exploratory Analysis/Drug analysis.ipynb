```python
# Data exploration
import pandas as pd

# Numerical
import numpy as np
import data_utility

#Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import the functions from their corresponding files
from NYC_GetCleaned_PrecinctData import getPrecinctData
from data_utility import filterData,getCount
from NYC_GetCleaned_TotalPopulation import getMeanPopulation
```

```python
 crimes_original = getPrecinctData()
crimes_original
```

```python
data = filterData(crimes_original,'OFNS_DESC','DANGEROUS DRUGS')
data.head()
```

```python
boroList = data['BORO_NM'].unique().tolist()
```

```python
labels = data['PD_DESC'].unique().tolist()
```

```python
data['PD_DESC'].value_counts(dropna=False)
```

```python
pop=[]
popPercent=[]
def getPopulation(borolist):
    pop.clear()
    popPercent.clear()
    i=1
    for boro in borolist:
        popValue = getMeanPopulation(boro,i)
        pop.append(popValue)
        i = i+1
    popsum = int(sum(pop))
    for value in pop:
        popPercentValue = value*100/popsum
        popPercent.append(round(popPercentValue,2))
    return popPercent    
    
```

```python
getPopulation(boroList)
```

```python
drugsList = []
drugsSumList = []
def getDrugDistribution(boroList):
    drugsList.clear()
    drugsSumList.clear()
    for boro in boroList:
        temp = filterData(data,'BORO_NM', boro)
        tempCount = len(temp['PD_DESC'].tolist())
        drugsList.append(tempCount)
    drugsum = sum(drugsList)
    for i in drugsList:
         percent= i*100/drugsum
         drugsSumList.append(round(percent,2))
    return drugsSumList
    
```

```python
getDrugDistribution(boroList)
```

```python
ratioList = []
def getvalueByPop(popPercent,listValue):
    ratioList.clear()
    for i in range(5):
         ratio = listValue[i]/popPercent[i]
         ratioList.append(ratio)
    return ratioList
```

```python
precinct_boro = pd.DataFrame({'Boroughs':boroList,'No. of Drugs related Crime(%)': getDrugDistribution(boroList),'Ratio by Population':getvalueByPop(popPercent,drugsSumList)})
precinct_boro
```

```python
a4_dims = (14, 8)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(x='Boroughs', y='No. of Drugs related Crime(%)', data=precinct_boro, ax=ax)
plt.ylabel('No. of Drugs related Crime(%)')
plt.title('Drug related Crime per population')
plt.show()
```

```python
a4_dims = (14, 8)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(x='Boroughs', y='Ratio by Population', data=precinct_boro, ax=ax)
plt.ylabel('Ratio by Population')
plt.title('Drug related Crime per population')
plt.show()
```
