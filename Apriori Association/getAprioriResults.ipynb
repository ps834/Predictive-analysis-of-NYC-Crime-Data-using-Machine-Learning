```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori

store_data = pd.read_csv('C:\\Users\\Pooja Suvarna\\Desktop\\made.csv', header=None) 
store_data.head() 
```

```python
records = []  
for i in range(0, 4):  
    records.append([str(store_data.values[i,j]) for j in range(0, 5098)])
```

```python
association_rules = apriori(records, min_support=0.5, min_confidence=0.3, min_lift=0, min_length=0)  
association_results = list(association_rules)  
```
